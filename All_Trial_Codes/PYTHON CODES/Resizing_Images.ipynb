```python
from PIL import Image
import os

# Kaynak ve hedef klasör yolları
kaynak_klasoru = "/content/gdrive/MyDrive/AliBaki_TURKOZ_Engineering_Project/Data_Sets/Grafikler/Sağlıklı Grafik"
hedef_klasoru = "/content/gdrive/MyDrive/AliBaki_TURKOZ_Engineering_Project/Data_Sets/Grafikler/Sağlıklı Grafik 224x224"

# Hedef boyutlar
hedef_genislik = 224
hedef_yukseklik = 224

# Kaynak klasördeki her bir görüntüyü işleme
for dosya_adi in os.listdir(kaynak_klasoru):
    dosya_yolu = os.path.join(kaynak_klasoru, dosya_adi)

    # Dosya uzantısını kontrol etme
    if dosya_adi.lower().endswith(".png"):  # Sadece PNG dosyalarını işleme
        # Görüntüyü yükleme
        img = Image.open(dosya_yolu)

        # Görüntünün boyutlarını alınması
        genislik, yukseklik = img.size

        # Boyutları 256x256 boyutuna getirme
        oran = min(hedef_genislik/genislik, hedef_yukseklik/yukseklik)
        yeni_genislik = int(genislik * oran)
        yeni_yukseklik = int(yukseklik * oran)
        img_resized = img.resize((yeni_genislik, yeni_yukseklik), Image.LANCZOS)

        # Yeni bir görüntü oluşturma ve hedef boyutunda beyaz arka plan ekleme
        yeni_img = Image.new("RGB", (hedef_genislik, hedef_yukseklik), (255, 255, 255))
        sol_ust_x = (hedef_genislik - yeni_genislik) // 2
        sol_ust_y = (hedef_yukseklik - yeni_yukseklik) // 2
        yeni_img.paste(img_resized, (sol_ust_x, sol_ust_y))

        # Yeni görüntüyü kaydetme
        kaydedilen_dosya_yolu = os.path.join(hedef_klasoru, dosya_adi)
        yeni_img.save(kaydedilen_dosya_yolu)
    else:
        print(f"{dosya_adi} dosyası bir PNG dosyası değil, işlenmedi.")

print("Tüm görüntüler başarıyla boyutlandırıldı ve kaydedildi.")
```
